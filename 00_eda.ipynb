```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('BRCA_data.csv')
```

```python
df
```

```python
sample_name = df.iloc[:,0].copy()    # sample index
genes_df = df.iloc[:,1:354].copy()   # X : gene expression features
clinical_df = df.iloc[:,354:].copy() # y : labels
```

```python
genes_df.describe()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(genes_df.T, cmap="viridis", cbar_kws={'label': 'Expression Level'}, xticklabels=False)
plt.title("Gene Expression Across Tumor Samples")
plt.ylabel("Genes")
plt.xlabel("Samples")
plt.savefig("docs/brca_heatmap.png")
plt.show()
```
